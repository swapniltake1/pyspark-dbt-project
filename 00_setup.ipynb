```python
# Create Catalog
spark.sql("""
CREATE CATALOG IF NOT EXISTS airline_catalog
""")
```

```python
# Use Catalog
spark.sql("USE CATALOG airline_catalog")

# Create Schema
spark.sql("""
CREATE SCHEMA IF NOT EXISTS reservation_schema
""")

# Set Active Schema
spark.sql("USE reservation_schema")

print("Schema Created & Activated")
```

```python
# Create Volume for Raw & Processed Data
spark.sql("""
CREATE VOLUME IF NOT EXISTS reservation_schema.airline_volume
""")

print("Volume Created Successfully")
```

```python
# List Schemas
display(spark.sql("SHOW SCHEMAS IN airline_catalog"))

# List Volumes
display(spark.sql("SHOW VOLUMES IN reservation_schema"))
```
